# Stemming

```python
from nltk.stem import PorterStemmer
```

```python
words = ["program", "programs", "programer", "programing", "programers", "go", "goes", "going", "gone"]
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
  print(word+ "----->" +stemming.stem(word))
```

```python
word1 =stemming.stem("congratulations")
```

```python
word1
```

```python
from nltk.stem import RegexpStemmer
```

```python
stemmer = RegexpStemmer("ing$|s$|e$|able$|ly$")
```

```python
word2 = stemmer.stem("consistently")
```

```python
word2
```
